```python
import sys, os

fromdir = "/media/ronit/New Volume/CODDING/PROJECTS/ionia-next"

def print_structure(path, indent=""):
    """Recursively prints the file structure, skipping .git, .next, and node_modules."""
    for root, dirs, files in os.walk(path):
        # Skip .git, .next, and node_modules directories
        dirs[:] = [d for d in dirs if d not in ('.git', '.next', 'node_modules')]

        # Get relative path from the root directory
        rel_root = os.path.relpath(root, fromdir)
        if rel_root == ".":
            rel_root = ""  # Ensure root directory is printed without extra "/"
        
        # Print the current directory if it's not the base directory
        if rel_root:
            print(f"{indent}{rel_root}/")
            indent += "    "  # Add indentation for subdirectories
        
        # Print files inside the current directory
        for file in files:
            file_path = os.path.join(root, file)
            rel_file = os.path.relpath(file_path, fromdir)
            print(f"{indent}├── {rel_file}")

# Start from the given directory

```
